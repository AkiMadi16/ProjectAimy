```python
# Import PySpark SQL functions
from pyspark.sql import functions as F

# Import Window functions
from pyspark.sql.window import Window
```

```python
`functions as F`

Provides built-in PySpark functions used for:

- Selecting columns
- Creating calculated columns
- Filtering records
- Aggregations
- Date transformations
- String operations

Using the alias `F` makes the code shorter and easier to read.

Example:

F.col("AttendeeId")
F.when(...)
F.count(...)

### Window

A Window defines how rows should be grouped and ordered before applying calculations such as:

- ROW_NUMBER()
- RANK()
- LEAD()
- LAG()

In this project, Window functions are used to identify the most recent active record for each business entity while preserving historical data.


```
